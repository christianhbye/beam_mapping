```python
import astropy
import croissant as cro
import croissant.jax as crojax
from functools import partial
from healpy import get_nside, projview
import jax
import jax.numpy as jnp
import matplotlib.pyplot as plt
from pygdsm import GlobalSkyModel16 as GSM16
import s2fft

%matplotlib inline
```

## Beam

```python
def plot_polar(beam, title="Beam Pattern"):
    fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
    ax.plot(theta, beam, c="C0")
    ax.plot(-theta, beam, c="C0")
    ax.set_rmax(1.03)
    ax.set_rticks([0.25, 0.5, 0.75])
    ax.set_rlabel_position(-15)  # Move radial labels away from plotted line
    ax.grid(True)
    ax.set_theta_zero_location("N")
    ax.set_title(title, va="bottom")
    return ax

# we use a coiled dipole that acts as a half-wavelength dipole at each frequency
# radiation pattern is given in Balanis, page 182 (section 4.6)
theta = jnp.linspace(0, jnp.pi, num=181, endpoint=True)  # 1 deg spacing, 0 to pi
beamZ = jnp.sin(theta)**3
_ = plot_polar(beamZ)

# the beam above is for a vertical dipole, aligned with the z-axis. we use two dipoles along x and y
phi = jnp.linspace(0, 2*jnp.pi, num=360, endpoint=False)
phi, theta = jnp.meshgrid(phi, theta, sparse=True)
x = jnp.sin(theta) * jnp.cos(phi)
y = jnp.sin(theta) * jnp.sin(phi)
z = jnp.cos(theta)

beamX = (y**2 + z**2) ** (3/2)
beamY = (x**2 + z**2) ** (3/2)

phi_plot = 0
ax = plot_polar(beamX[:, phi_plot], title=f"Beam at $\\phi={phi_plot}$ deg")
ax.plot(theta, beamY[:, phi_plot], c="C1")
ax.plot(-theta, beamY[:, phi_plot], c="C1")
plt.show()
```

```python
beam = 1/2 * (beamX + beamY)
horizon_mask = jnp.where(theta < jnp.pi/2, 1, 0)
plt.figure()
plt.imshow(beam*horizon_mask, aspect="auto")
plt.colorbar()
plt.ylabel("$\\theta$")
plt.xlabel("$\\phi$")
plt.show()
```

```python
# simulation parameters
world = "earth"
lmax = 128  # maximal harmonic band limit given sampling of beam, inclusive
freq = jnp.arange(50, 101)  # 50-101 MHz
time = astropy.time.Time("2025-12-01 09:00:00")  # time at the beginning of the simulation
loc = astropy.coordinates.EarthLocation(lon=-113, lat=39)  # location of telescope
topo = astropy.coordinates.AltAz(obstime=time, location=loc) # coordinate frame of telescope
# 240 bins in a sidereal day on the moon
ntimes = 240
dt = cro.constants.sidereal_day[world] / ntimes
phases = crojax.simulator.rot_alm_z(lmax, ntimes, dt, world=world)

# get the euler angles and wigner d functions for the coordinate transforms
eul_topo, dl_topo = crojax.rotations.generate_euler_dl(lmax, topo, "fk5")  # beam transform, from topocentric to mcmf
eul_gal, dl_gal = crojax.rotations.generate_euler_dl(lmax, "galactic", "fk5")  # sky transform, from galactic to mcmf

topo2eq = partial(s2fft.utils.rotation.rotate_flms, L=lmax+1, rotation=eul_topo, dl_array=dl_topo)
gal2eq = partial(s2fft.utils.rotation.rotate_flms, L=lmax+1, rotation=eul_gal, dl_array=dl_gal)
```

```python
# define the map2alm transform for the beam
beam2alm = partial(s2fft.forward_jax, L=lmax+1, spin=0, nside=None, sampling="mwss", reality=True)
power = crojax.alm.total_power(beam2alm(beam), lmax)
# use vmap to vectorize frequency axis
beam = jnp.repeat(beam[None], freq.size, axis=0)  # achromatic beam for now
beam_alm = jax.vmap(beam2alm)(beam*horizon_mask)

# normalization for visibilities
fsky = crojax.alm.total_power(beam_alm, lmax) / power  # part of beam pointing to sky
fground = 1 - fsky
print(fsky, fground)
```

```python
gsm = GSM16(freq_unit="MHz", data_unit="TRJ", resolution="lo")
sky_map = gsm.generate(freq)
ix = 50
projview(m=sky_map[ix], title=f"GSM at {freq[ix]} MHz", min=0, max=1e4)
plt.show()
```

```python
# define the map2alm transform for the sky
nside = get_nside(sky_map[0])
sky2alm = partial(s2fft.forward_jax, L=lmax+1, spin=0, nside=nside, sampling="healpix", reality=True)
sky_alm = jax.vmap(sky2alm)(sky_map)
```

```python
# coordinate transform
beam_alm = jax.vmap(topo2eq)(beam_alm)
sky_alm = jax.vmap(gal2eq)(sky_alm)
```

```python
# beam and sky as seen by simulator after transformed back to healpix
beam_reconstructed = s2fft.inverse_jax(beam_alm[ix], L=lmax+1, spin=0, nside=nside, sampling="healpix", reality=True)
sky_reconstructed = s2fft.inverse_jax(sky_alm[ix], L=lmax+1, spin=0, nside=nside, sampling="healpix", reality=True).real

projview(m=beam_reconstructed.real, title=f"Reconstructed beam at {freq[ix]} MHz")
projview(m=sky_reconstructed, title=f"Reconstructed sky at {freq[ix]} MHz")
plt.show()
```

```python
T_sky = crojax.simulator.convolve(beam_alm, sky_alm, phases).real / power
cmin = 0
cmax = 7000
plt.figure()
im = plt.imshow(T_sky, aspect="auto")
im.set_clim(cmin, cmax)
plt.colorbar()
plt.show()
```

We take the ground contribution to be a blackbody component of temperature 300 K and a uniform reflection of the part of the sky above the horizon, times some coupling coefficient. The scattered part is therefore simply the convolution of the beam with a sky that is 0 above the horizon and a constant temperature below the horizon. However, we invoke a trick: using the commutatitvity of multiplication, we multiply this horizon mask into the beam instead of the sky. Thus the beam can be thought of as 0 above the horizon and with some complicated pattern below the horizon, whereas the sky is simply a monpole. Only the monopole component of the beam is sensitive to this sky; and so we really only pick up a factor of $f_{\textrm{ground}}$ multiplied into the blackbody ground. Now, the temperature of the scattered radiation is:
$$T_{\textrm{scatt}} = \int_{\textrm{sky}} T_{\textrm{sky}} \mathrm{d}\Omega = \int_{4 \pi} T_{\textrm{sky}} H \mathrm{d}\Omega$$

Here, $H$ is a indicator function, that is 1 above the horizon and 0 below the horizon. Now, we recognize this simply as another convolution. Thus, we simply get the spherical harmonics decomposition of the indicator function and convolve it with the sky using the simulator. This even gives us time dependence for free!

The temperature of the ground is simply:
$$T_{\textrm{ground}} = 300 \textrm{K} + \epsilon T_{\textrm{scatt}}$$

and the total antenna temperature becomes:
$$T_{\textrm{ant}} = f_{\textrm{sky}} T_{\textrm{sky}} + f_{\textrm{ground}} T_{\textrm{ground}}$$

We note that the ground contribution appears at a delay set by the light travel time from the antenna platform to the ground and back up. This corresponds to a phase shift in the Fourier transform of the spectra, see below.

```python
H_alm = beam2alm(jnp.repeat(horizon, phi.size, axis=1))  # indicator function / horizon mask
H_alm = topo2eq(H_alm)
T_scatt = crojax.simulator.convolve(H_alm[None], sky_alm, phases).real
eps = 1e-3  # coupling coefficient
T_ground = 300 + eps * T_scatt
c = 299792458  # m/s
dly = 2 * 100 / c  # 100m to ground, factor of 2 for back and forth
```

```python
plt.figure()
im = plt.imshow(T_ground, aspect="auto")
plt.colorbar()
plt.show()
```

```python
_T_sky = jnp.fft.fft(fsky * T_sky, axis=1)  # sky contribution only
_T_ground = jnp.fft.fft(fground * T_ground, axis=1) * jnp.exp(-2*jnp.pi*dly*freq)
_T_ant = _T_sky + _T_ground
T_ant = jnp.fft.ifft(_T_ant).real  # check fft and ifft order
plt.figure()
plt.imshow(T_ant, aspect="auto")
plt.colorbar()
plt.show()
```

```python
tau = jnp.fft.fftfreq(freq.size, d=1e6) * 1e9  # nanoseconds
plt.figure()
plt.plot(tau, _T_sky[0])
plt.plot(tau, _T_ground[0])
plt.plot(tau, _T_ant[0])
plt.show()
```
